```python
#流星のベクトル求める
#part2;上のsectionで求めた流星の2点のra,decをaz,elに変換
import pandas as pd
import math
import numpy as np
from decimal import *

df1 = pd.read_csv("Kisodata_85mm.csv")
df2 = pd.read_csv("Kisodata_85mm.csv")
# Constant
longitude_deg = 137.62535
latitude_deg = 35.7973


df1["year_tomoe"] = df1["Y(JST)"].astype(float)
df1["month_tomoe"] = df1["M(JST)"].astype(float)
df1["day_tomoe"] = df1["D(UT)"].astype(float)
df1["hh[UT]_tomoe"] = df1["H(UT)"].astype(float)
df1["mm_tomoe"] = df1["M"].astype(float)
df1["ss_tomoe"] = df1["S"].astype(float)


floor_1 = 365.25*(df1["year_tomoe"] + 4716.0)
floor_2 = (30.6001*(df1["month_tomoe"] + 1.0) + 2.0)
floor_3 = df1["year_tomoe"]/100.0
floor_4 = (df1["year_tomoe"]/100.0)/4
floor_1 = floor_1.astype(int)
floor_2 = floor_2.astype(int)
floor_3 = floor_3.astype(int)
floor_4 = floor_4.astype(int)


df1["JD"] = floor_1 + floor_2 - floor_3 + floor_4 +df1["day_tomoe"] - 1524.5 + \
            (df1["hh[UT]_tomoe"] + df1["mm_tomoe"]/60 + df1["ss_tomoe"]/3600)/24

D = df1["JD"] - 2451543.5
w = 282.9404 + 4.70935e-5*D
M = (356.0470 + 0.9856002585*D)%360
L = w + M
GMST0 = ((L + 180)%360)/15
UT_hour = df1["hh[UT]_tomoe"] + df1["mm_tomoe"]/60 + df1["ss_tomoe"]/3600
SiderealTime = GMST0 + UT_hour + longitude_deg/15

HourAngle_1 = (SiderealTime*15 - df1[" Ra1"])
A_1 = np.cos(np.radians(HourAngle_1))*np.cos(np.radians(df1[" Dec1"]))*np.cos(np.radians(90 - latitude_deg)) - \
    np.sin(np.radians(df1[" Dec1"]))*np.sin(np.radians(90 - latitude_deg))
B_1 = np.sin(np.radians(HourAngle_1))*np.cos(np.radians(df1[" Dec1"]))
C_1 = np.cos(np.radians(HourAngle_1))*np.cos(np.radians(df1[" Dec1"]))*np.sin(np.radians(90 - latitude_deg)) + \
    np.sin(np.radians(df1[" Dec1"]))*np.cos(np.radians(90 - latitude_deg))

HourAngle_2 = (SiderealTime*15 - df1[" Ra2"])
A_2 = np.cos(np.radians(HourAngle_2))*np.cos(np.radians(df1[" Dec2"]))*np.cos(np.radians(90 - latitude_deg)) - \
    np.sin(np.radians(df1[" Dec2"]))*np.sin(np.radians(90 - latitude_deg))
B_2 = np.sin(np.radians(HourAngle_2))*np.cos(np.radians(df1[" Dec2"]))
C_2 = np.cos(np.radians(HourAngle_2))*np.cos(np.radians(df1[" Dec2"]))*np.sin(np.radians(90 - latitude_deg)) + \
    np.sin(np.radians(df1[" Dec2"]))*np.cos(np.radians(90 - latitude_deg))

az_1 = (np.arctan(B_1/A_1)*180/np.pi) + 180
el_1 = np.arcsin(C_1)*180/np.pi

az_2 = (np.arctan(B_2/A_2)*180/np.pi) + 180
el_2 = np.arcsin(C_2)*180/np.pi

#azel_1 = np.hstack((az_1, el_1))
#azel_2 = np.hstack((az_2, el_2))

columns_1 = ["az1[deg]"]
columns_2 = ["el1[deg]"]
columns_3 = ["az2[deg]"]
columns_4 = ["el2[deg]"]
az_1 = pd.DataFrame(data = az_1, columns = columns_1)
el_1 = pd.DataFrame(data = el_1, columns = columns_2)
az_2 = pd.DataFrame(data = az_2, columns = columns_3)
el_2 = pd.DataFrame(data = el_2, columns = columns_4)

df2 = pd.concat([df2, az_1], axis=1)
df2 = pd.concat([df2, el_1], axis=1)
df2 = pd.concat([df2, az_2], axis=1)
df2 = pd.concat([df2, el_2], axis=1)

df2.to_csv("kiso_data_azel_85mm.csv")

```
