```python
#　Observation time
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from IPython.display import clear_output
from matplotlib.backends.backend_pdf import PdfPages

df_mu = pd.read_csv("../MU_data_2009-2015.csv")
day = df_mu["DD"].values
hour = df_mu["hh"].values
minute = df_mu["mm"].values
time = day*24.*60. + hour*60. + minute

### total observation time [min]
total_observation_time = []
mjd = df_mu["MJD"].values
dif_mjd = mjd
for i in range(len(mjd)-1):
    dif_mjd[i] = mjd[i+1] - mjd[i]
separate_index = np.where(dif_mjd > 0.02) #about 30min
separate_index = list(separate_index)
separate_index = np.insert(separate_index, 0 , 0)

for j in range(len(separate_index)-1):
    observation_time = time[separate_index[j+1]] - time[separate_index[j]+1]
    total_observation_time = np.hstack((total_observation_time, observation_time))

total_observation_time = np.sum(total_observation_time)
total_observation_time_hour = total_observation_time/60.
print "observation time is",total_observation_time, "[min]"
print "observation time is",total_observation_time_hour, "[hour]"
print "Complete!"
```

```python
#　population index
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from IPython.display import clear_output
from matplotlib.backends.backend_pdf import PdfPages

collection_area = pd.read_csv("collection_area.csv")
df_mu = pd.read_csv("../MU_data_2009-2015.csv")
array_mu = df_mu.values
total_observation_time_hour = 845.8 #[hour]
collection_area_uncorrected = np.pi*(100*np.tan(np.deg2rad(3.)))**2

def linear_fit(x, a, b):
    return b*a**(x)

#RCSが空, RSC>20.0 のデータを削除
#RCS = -4.58*MAG+16.92 ⇨ MAG = RCS/(-4.58) + 16.92/4.58
rcs = df_mu["RCS"].values
zenith_distance = df_mu["Ze_(deg)"].values
rcs_unicode = rcs.astype(unicode)
nan = np.array(["nan"])
nan = nan.reshape(1)
nan = nan.astype(unicode)
delete_index = np.where((rcs_unicode == nan) | (rcs > 20.0) | (zenith_distance > 90.0))
array_mu = np.delete(array_mu, delete_index, 0)
rcs = np.delete(rcs, delete_index, 0)
zenith_distance = np.delete(zenith_distance, delete_index, 0)

################################################################

zenith_correction = 1./np.cos(np.deg2rad(zenith_distance))
rcs_bins = collection_area["rcs_dBsm"].values
collection_area_table = collection_area["collection_area"].values
rcs_axis = np.ones(len(rcs_bins) -1)
number = np.ones(len(rcs_bins) -1)
collection_area = np.ones(len(rcs_bins) -1)
cumulative_flux = np.ones(len(rcs_bins) -1)
cumulative_flux_uncorrected = np.ones(len(rcs_bins) -1)

for i in range(len(rcs_bins)-1):
    index = np.array(np.where((rcs > rcs_bins[i]) & (rcs < rcs_bins[i+1])))
    num = zenith_correction[index]
    collection_area[i] = (collection_area_table[i] + collection_area_table[i+1])/2.
    rcs_axis[i] = (rcs_bins[i] + rcs_bins[i+1])/2.
    number[i] = np.sum(num)
    cumulative_flux[i] = number[i]/collection_area[i]/total_observation_time_hour
    cumulative_flux_uncorrected[i] = number[i]/collection_area_uncorrected/total_observation_time_hour

number = number[::-1]
cumulative_number = np.cumsum(number)
cumulative_flux_err = np.sqrt(cumulative_number)/collection_area/total_observation_time_hour
cumulative_flux = cumulative_flux[::-1]
cumulative_flux = np.cumsum(cumulative_flux)
cumulative_flux_uncorrected = cumulative_flux_uncorrected[::-1]
cumulative_flux_uncorrected = np.cumsum(cumulative_flux_uncorrected)
rcs_axis = rcs_axis[::-1]
mag_axis = rcs_axis/(-4.58) + 16.92/4.58

################################################################

param, cov = curve_fit(linear_fit, mag_axis[5:120], cumulative_flux[5:120],sigma=cumulative_flux_err[5:120])
y_fit = param[1]*param[0]**(mag_axis)

cumulative_flux_max = cumulative_flux + cumulative_flux_err
cumulative_flux_min = cumulative_flux - cumulative_flux_err

fig = plt.figure(figsize=(10,10))
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams["mathtext.fontset"] = 'cm'
plt.rcParams["font.size"] = 25
plt.plot(mag_axis,cumulative_flux,color="darkblue",linewidth=4,label='Corrected')
plt.plot(mag_axis,cumulative_flux_uncorrected,color="red",linewidth=4,alpha=0.3,label='Uncorrected')
plt.fill_between(mag_axis,cumulative_flux_min,cumulative_flux_max,alpha=0.2)
plt.plot(mag_axis,y_fit,linewidth=2,linestyle="dashdot",color="darkblue")
plt.grid(which='major',alpha=0.2,color='black',linestyle='-')
plt.yscale("log")
plt.xlabel("Vband Absolute Magnitude")
plt.ylabel("Log Cumulative Flux [$km^{-2} hour^{-1}$]")
plt.text(-1,8,"Population Index r = 2.69$\pm$ 0.0185",color='black')
plt.text(-1,2,"Fitting range [0:8.5]",color='black')
plt.legend(loc='upper left')
pdf = PdfPages("d_img/population_index.pdf")
pdf.savefig()
pdf.close()
plt.savefig("d_img/population_index.png")
plt.show()

clear_output()
print "population index", param, "error", np.sqrt(np.diag(cov))
print "fitting range is", mag_axis[9], ":", mag_axis[120]
print "Complete!"
```
